```python
import pandas as pd

messages= pd.read_csv(r'C:\Users\Cynthia\Documents\NLP Learnings & Implemetation\NLP-Learnings-Basic-Implementation\Input Dataset\SMSSpamCollection'
                      ,sep='\t',names=['label','message'])
```

### Data Cleaning & Pre Processing ###

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i]) #Replace all non-alphabetical characters with space
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

### Bag of Words ###

```python
from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer(max_features=2500,binary=True) # we are taking only top 2500 features which are most frequently occuring
X= cv.fit_transform(corpus).toarray()
```
